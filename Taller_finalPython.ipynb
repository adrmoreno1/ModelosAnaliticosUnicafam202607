```python
#6. FizzBuzz clásico: Para un número n, imprime: "FizzBuzz" si es múltiplo de 3 y de 5; "Fizz" si es múltiplo solo de 3;
# "Buzz" si es múltiplo solo de 5; el número si no cumple ninguna. Prueba con n = 15, n = 9, n = 10 y n = 7.

def fizzbuzz(numero):
    if numero % 3 == 0 and numero % 5 == 0:
        print("FizzBuzz")
    elif numero % 3 == 0:
        print("Fizz")
    elif numero % 5 == 0:
        print("Buzz")
    else:
        print(numero)

fizzbuzz(15)  # FizzBuzz
fizzbuzz(9)   # Fizz
fizzbuzz(10)  # Buzz
fizzbuzz(7)   # 7
```

```python
#3. Recorrer una lista: Dada precios = [12000, 45000, 8500, 67000, 3200], usa un ciclo para: (1) imprimir cada precio con
#formato Item 1: $12,000, (2) imprimir al final cuántos artículos cuestan más de $10,000.

precios =  [12000, 45000, 8500, 67000, 3200]
for i in precios:
    print(f"El precio es ${i:,}")


#Cuantos articulos cuestan más de $10,000
precios =  [12000, 45000, 8500, 67000, 3200]
cantidad = 0
for i in precios:
    if (i > 10000):
      cantidad += 1

print(f"{cantidad} articulos cuestan mas de $10,000")
```

```python
#4. While con condición: Un banco tiene saldo = 1000.
#Cada mes retiras 150 y se cobran 20 de comisión. Usa un ciclo while para simular cuántos meses pasan antes de que el saldo llegue a cero o menos. Imprime el saldo al final de cada mes.

saldo = 1000
mes   = 0

while saldo > 0:
    saldo = saldo - 150
    saldo = saldo - 20
    mes   += 1
    print(f"Mes {mes}: saldo ${saldo}")

print(f"El dinero se acaba en  {mes} meses")
```

```python
#5. break y continue: Dada la lista numeros = [3, 7, -2, 15, 4, -8, 9, 1],
#escribe un ciclo que: (1) saltee los números negativos con continue, (2) se detenga con break al encontrar un número mayor a 10, (3) imprima los números que procesa.

numeros = [3, 7, -2, 15, 4, -8, 9, 1]

for i in numeros:
  if (i < 0):
    continue
  elif (i > 10):
    break;
  print(f"El número procesado es: {i}")
```
