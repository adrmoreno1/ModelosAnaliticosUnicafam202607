1. Agregar un dato nuevo
2. Actualizar un dato existente
3. Eliminar un dato con del
4. Eliminar un dato con .pop()
5. Consultar un dato con .get()
6. Ver el perfil completo

```python
# datos
perfil = {
    "nombre": "Adriana Maria Moreno Lopez",
    "empresa": "Universitas",
    "hobbies": ["viajar", "comer rico", "correr", "ver series"],
    "altura": "1.65 m",
    "peso": "55 kg"
}

# Menú
print("===== MENÚ DE MI PERFIL =====")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato existente")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop()")
print("5. Consultar un dato con get()")
print("6. Ver el perfil completo")

opcion = input("Seleccione una opción (1-6): ")

# Opción 1
if opcion == "1":
    clave = input("Ingrese la nueva clave: ")
    valor = input("Ingrese el valor: ")
    perfil[clave] = valor
    print("\nDato agregado correctamente.")
    print(perfil)

# Opción 2
elif opcion == "2":
    clave = input("Ingrese la clave que desea actualizar: ")

    if clave in perfil:
        nuevo_valor = input("Ingrese el nuevo valor: ")

        if clave == "hobbies":
            perfil[clave] = nuevo_valor.split(",")
        else:
            perfil[clave] = nuevo_valor

        print("\nDato actualizado correctamente.")
    else:
        print("\nLa clave no existe.")

    print(perfil)

# Opción 3
elif opcion == "3":
    clave = input("Ingrese la clave que desea eliminar: ")

    if clave in perfil:
        del perfil[clave]
        print("\nDato eliminado correctamente.")
    else:
        print("\nLa clave no existe.")

    print(perfil)

# Opción 4
elif opcion == "4":
    clave = input("Ingrese la clave que desea eliminar: ")

    if clave in perfil:
        valor_eliminado = perfil.pop(clave)
        print("\nDato eliminado correctamente.")
        print("Valor eliminado:", valor_eliminado)
    else:
        print("\nLa clave no existe.")

    print(perfil)

# Opción 5
elif opcion == "5":
    clave = input("Ingrese la clave que desea consultar: ")
    valor = perfil.get(clave, "No disponible")
    print("\nValor encontrado:", valor)

# Opción 6
elif opcion == "6":
    print("\n***** PERFIL COMPLETO *****")
    print(perfil)

# Opción inválida
else:
    print("Opción inválida.")
```
