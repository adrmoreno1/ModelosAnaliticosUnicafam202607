##Listas

```python
#Una lista vacia
mi_lista = []
#Lista con varios elementos
mi_lista = [1,2,3,"4",5.0, True]
print(mi_lista[2])
```

```python
#Cambio en la lista
mi_lista[0]='Adriana'
print(mi_lista)
```

```python
#Agregar elementos a una lista
mi_lista.append("Moreno")
print(mi_lista)
```

```python
#Agregar elementos a una posición especifica
mi_lista.insert(3, "Bogotá")
print(mi_lista)
```

```python
#Agregar varios elementos al final de la lista
mi_lista.extend([7,False])
print(mi_lista)
```

```python
#Eliminación de objetos especificos
mi_lista.remove(True)
print(mi_lista)
```

```python
print("ACTUALIZACIÓN DE DATOS")
valor_buscar = input("Por favor ingrese el nombre para actualizar: ")
valor_reemplazar = input("Por favor ingrese el nombre para reemplazar: ")
mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
print("Cambio ejecutado imprimiendo resultado")
print(mi_lista)
```

```python
print("ELIMINACIÓN DE DATOS")
valor_eliminar = input("Por favor ingrese el nombre para eliminar: ")
mi_lista.remove(valor_eliminar)
print("Cambio ejecutado imprimiendo resultado")
print(mi_lista)
```

##Diccionarios

```python
mi_diccionario = {
    "nombre" :  "Adriana",
    "ciudad" : "Bogotá",
    "Empresa": "Universitas",
}
```

```python
#Acceder a los elementos del diccionario
print(mi_diccionario["nombre"])
```

```python
#Modificar un valor en el diccionario
mi_diccionario["nombre"]= "Adriana Maria"
print(mi_diccionario)
```

```python
#Añadir un valor en el diccionario
mi_diccionario["Profesión"]= "Ingeniera de sistemas"
print(mi_diccionario)
```

```python
#Eliminar un elemento del diccionario
del mi_diccionario["ciudad"]
print(mi_diccionario)
```

##CONDICIONALES

```python
#Crea un programa que determine si una persona es mayor de edad o menor de edad
print("---- VERIFICACIÓN DE EDAD---")
edad=int(input("Ingresa tu edad: "))
if (edad < 18):
  print(f"Eres menor de edad, tienes: {edad}")
else:
  print(f"Eres mayor de edad, tienes: {edad}")
```
